```python
import pandas as pd 
```

```python
census = pd.read_csv('CENSUS DATA ACSDP5Y2018.DP05_data_with_overlays_2020-03-26T100029_Edited.csv')
```

```python
covid = pd.read_csv('COVID-19 Data JHU 04-13-2020.csv')
```

```python
unemployment = pd.read_csv('US State 2000 to present Unemployment Data 14Apr2020.csv')
```

```python
census.head()
```

```python
census_new = census.rename(columns={'Geographic Area Name': 'State'},)
```

```python
census_new.head()
```

```python
covid.head()
```

```python
covid_new = covid.rename(columns={'Province_State': 'State'},)
```

```python
covid_new.head()
```

```python
covid_usa_only = covid_new[covid_new["Country_Region"] == 'US']
```

```python
covid_usa_only.head()
```

```python
#Grouping Covid cases by State
covid_usa_only.groupby('State').agg({'Confirmed':['sum']}).head()
```

```python
Confirmed_Covid_Sum = covid_usa_only.groupby('State').agg({'Confirmed':['sum']})
```

```python
Confirmed_Covid_Sum = Confirmed_Covid_Sum.rename(columns={'Confirmed': 'Covid_Confirmed'},)
```

```python
Confirmed_Covid_Sum.head()
```

```python
Death_Covid_Sum = covid_usa_only.groupby('State').agg({'Deaths':['sum']})
```

```python
Death_Covid_Sum = Death_Covid_Sum.rename(columns={'Deaths': 'Covid_Deaths'},)
```

```python
Death_Covid_Sum.head()
```

```python
Recovered_Covid_Sum = covid_usa_only.groupby('State').agg({'Recovered':['sum']})
```

```python
Recovered_Covid_Sum = Recovered_Covid_Sum.rename(columns={'Recovered': 'Covid_Recovered'},)
```

```python
Recovered_Covid_Sum
```

```python
census_covid_v1 = pd.merge(Confirmed_Covid_Sum, census_new, on='State', how='inner')
```

```python
census_covid_v1
```

```python
census_covid_v2 = pd.merge(Death_Covid_Sum, census_covid_v1, on='State', how='inner')
```

```python
census_covid_v2
```

```python
census_covid_v3 = pd.merge(Recovered_Covid_Sum, census_covid_v2, on='State', how='inner')
```

```python
census_covid_v3
```

```python
census_covid_v3.to_csv('covid_census_merged.csv')
```

```python
covid_census_merged = pd.read_csv('covid_census_merged.csv')
```

```python
covid_census_merged.head()
```

```python
covid_census_merged = covid_census_merged.rename(columns={'(\'Covid_Recovered\', \'sum\')': 'Covid_Recovered'},)
```

```python
covid_census_merged.head()
```

```python
covid_census_merged = covid_census_merged.rename(columns={'(\'Covid_Deaths\', \'sum\')': 'Covid_Deaths'},)
```

```python
covid_census_merged.head()
```

```python
covid_census_merged = covid_census_merged.rename(columns={'(\'Covid_Confirmed\', \'sum\')': 'Covid_Confirmed'},)
```

```python
covid_census_merged.head()
```

```python
covid_census_merged = covid_census_merged.drop(['Unnamed: 0'], axis=1)
```

```python
covid_census_merged.head()
```

```python
def percent_population_confirmed_covid(row):
    confirmed_percent = row['Covid_Confirmed'] / row['Total population'] *100
    return confirmed_percent
```

```python
covid_census_merged['Covid_Confirmed_Percent_Total_Population'] = covid_census_merged.apply(percent_population_confirmed_covid, axis=1)
covid_census_merged.head()
```

```python
def percent_population_deaths_covid(row):
    deaths_percent = row['Covid_Deaths'] / row['Total population'] *100
    return deaths_percent
```

```python
covid_census_merged['Covid_Deaths_Percent_Total_Population'] = covid_census_merged.apply(percent_population_deaths_covid, axis=1)
covid_census_merged.head()
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15,5))
ax.scatter('Covid_Confirmed_Percent_Total_Population', 'Covid_Deaths_Percent_Total_Population', data=covid_census_merged);
ax.set_title('Covid Deaths % vs Covid Confirmed %')
ax.set_xlabel('Covid Confirmed Percent Total Population')
ax.set_ylabel('Covid Deaths Percent Total Population');
```

```python
unemployment = unemployment.loc[unemployment['Filed week ended'] == '3/28/2020']
```

```python
unemployment.head()
```

```python
unemployment['Initial Claims'] = unemployment['Initial Claims'].str.replace(',', '').astype(float)
```

```python
unemployment.head()
```

```python
unemployment['Continued Claims'] = unemployment['Continued Claims'].str.replace(',', '').astype(float)
```

```python
unemployment.head()
```

```python
def total_unemployment(row):
    total_initial_continued = row['Initial Claims'] + row['Continued Claims']
    return total_initial_continued
```

```python
unemployment['Inital plus Continued Claims'] = unemployment.apply(total_unemployment, axis=1)
unemployment.head()
```

```python
census_covid_unemployment = pd.merge(covid_census_merged, unemployment, on='State', how='inner')
```

```python
census_covid_unemployment.head()
```

```python
def unemployment_vs_total_population(row):
    unemployment_percent = row['Inital plus Continued Claims'] / row['Total population'] *100
    return unemployment_percent
```

```python
census_covid_unemployment['Unemployment_Percent'] = census_covid_unemployment.apply(unemployment_vs_total_population, axis=1)
census_covid_unemployment.head()
```

```python
plt.figure(figsize=(15, 15))

# You could do this one it matplotlib but it's much much easier in Seaborn,
# which makes it easy to add a color dimension.
sns.scatterplot('Unemployment_Percent', 'Covid_Confirmed_Percent_Total_Population', hue='State', data=census_covid_unemployment)
# Still easy to use the matplotlib interface to set labels/title.
plt.title('Confirmed Covid vs Unemployment')
plt.xlabel('Unemployment_Percent')
plt.ylabel('Covid_Confirmed_Percent_Total_Population');
```
