```python
import pandas as pd 
```

```python
census = pd.read_csv('CENSUS DATA ACSDP5Y2018.DP05_data_with_overlays_2020-03-26T100029_Edited.csv')
```

```python
covid = pd.read_csv('COVID-19 Data JHU 04-13-2020.csv')
```

```python
unemployment = pd.read_csv('US State 2000 to present Unemployment Data 14Apr2020.csv')
```

```python
census.head()
```

```python
census_new = census.rename(columns={'Geographic Area Name': 'State'},)
```

```python
census_new.head()
```

```python
unemployment.head()
```

```python
covid.head()
```

```python
covid_new = covid.rename(columns={'Province_State': 'State'},)
```

```python
covid_new.head()
```

```python
covid_usa_only = covid_new[covid_new["Country_Region"] == 'US']
```

```python
covid_usa_only.head()
```

```python
#Grouping Covid cases by State
covid_usa_only.groupby('State').agg({'Confirmed':['sum']}).head()
```

```python
Confirmed_Covid_Sum = covid_usa_only.groupby('State').agg({'Confirmed':['sum']})
```

```python
Confirmed_Covid_Sum.head()
```

```python
Death_Covid_Sum = covid_usa_only.groupby('State').agg({'Deaths':['sum']})
```

```python
Death_Covid_Sum.head()
```

```python
Recovered_Covid_Sum = covid_usa_only.groupby('State').agg({'Recovered':['sum']})
```

```python
Recovered_Covid_Sum.head()
```

```python
census_covid_v1 = pd.merge(Confirmed_Covid_Sum, census_new, on='State', how='inner')
```

```python
census_covid_v1.head()
```

```python
census_covid_v2 = pd.merge(Death_Covid_Sum, census_covid_v1, on='State', how='inner')
```

```python
census_covid_v2.head()
```

```python
census_covid_v3 = pd.merge(Recovered_Covid_Sum, census_covid_v2, on='State', how='inner')
```

```python
census_covid_v3.head()
```

```python
census_covid_rename_v1 = census_covid_v3.rename(columns={'(Recovered, sum)': 'Covid_Recovered'},)
```
